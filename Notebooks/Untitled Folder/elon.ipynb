# Test of zero-shot classification algorithm

## Preparation of model

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")
```

```python
tags = ["elon musk","illness", "neutral"]
test = tweets["text"].copy().head(5).tolist()
output = classifier(test, tags, multi_label=False)
print(output)
```

```python
raw_result = pd.DataFrame.from_dict(output)
raw_result
```
